# Chapter 3: Lists, Dictionaries, Tuples and Sets

Data structures are particular ways of sorting data to make some operation easier or more efficient. Certain data structures are more suited for solving particular problems, and are often associated with algorithms as well. We will be covering Data Structures and Algorithms in greater detail in a separate workshop.

In this Chapter, we will be covering some basic data structures, lists, dictionaries, tuples and sets, with particular emphasis on lists and dictionaries. The key properties and differences between these four data structures are as follows. 

*   **Lists:** A collection that is ordered (sequence), can be indexed and changeable *(or mutable)*. Allows duplicate members.
*   **Tuple:** A collection that is ordered (sequence), can be indexed and unchangeable *(or immutable)*. Allows duplicate members
*   **Set:** A collection that is unordered, unindexed and changeable. No duplicate members.
*   **Dictionary:** A collection that is unordered, can be indexed and changeable. No duplicate members (keys only).

Lets go through in detail what each data structure entails.


## 3.1 Lists

A lists can take on the form of something like this,

x=[1,2,3,4,5]

In this case, the square brackets [ ] help to denote the explicit items that belong in the list. The commas help to separate each element in the list.




```
x=[1,2,3,4,5]
print(x)
```

### 3.1.1 What can you have in a list?

You can have **any of the four data types**. You can even have a **list within a list**.

```
x=[9,1.2,'apple', True]

print(x)

y=[1,2,3,[4,5],6]

print(y)
```

### 3.1.2 Acessing the List

Indexing allows us to select a particular position in the list, and either **access the data at that position** or **assign data to that position**. Because lists are ordered, they are often indexed using numbers. Indexing are done using the **square brackets [ ]** as well.

Lists can be **indexed forwards**, i.e. starting from 0, in increments of 1, going from left to right or it can also be **indexed backwards**, i.e. starting from -1, decreasing by -1, from right to left.

<img src="https://drive.google.com/uc?id=1nW5tsex3uExCEE-aIR3AFeyFtdIX6yFd">






```
x=[9,1.2,'apple',True]
print(x)
print(x[0])
print(x[-4])

x[1]=25.2          #Assigning 25.2 to the index 1 of list x
print(x)
```

You can also do multiple indexing, otherwise known as **slicing**.

```
x=[9,1.2,'apple', True]
print(x[1:3])
```

Note that in this case the index syntax is as follows, [start:end (not including the end)].

We can also do consecutive indexing, which would be required if we have lists within a list.

```
y=[1,2,3,[4,5],6]

print(y[3][0])

print(y[1:5:2])
print(y[1,3,4])
```

Once again the indexing syntax starts from left to right, from outer list to inner list.

### 3.1.3 Lists Functions

* **len(list)**: Number of elements in the list (outermost)
* **min(list)**: Minimum element in the list
* **max(list)**: Maximum element in the list
* **sum(list)**: Sum up the elements in the list. Only works if all elements in the list are numeric
* element ** in** list: The in is not a function but rather a operator. Checks if a particular element is in the list (outermost).

```
x=[1,2,3,4,5]
print(len(x))
print(min(x))
print(max(x))
print(sum(x))
print(3 in x)
```

### 3.1.4 Lists Methods

Methods can be thought of as special functions that each object in Python has access to. In this case, a list would be our object.

The general syntax for using methods are **object.method()**, where the dot operator is used to call upond the methods associated with the particular object. 

If you are still confuse at this point, do not worry. Just know that to use methods, you do **object.method()**, and to use functions, you simply use the **function name**, i.e. **func()**.

Below are a few methods associated with Lists:
* lst.**append(ele)**: Adds an element to the end of the list
* lst.**extend(another_list)**: Adds a another list to the end of the target list
* lst.**pop(index)**: Removes an element, based on the given index
* lst.**insert(index,ele)**: Adds an element to the index position. Elements at and to the right of the index are shifted to the right.
* lst.**remove(ele)**: Removes the desired element (first occurence)
* lst.**sort()**: Sorts the list in ascending order. Works for both numerals and alphabets.
* lst.**reverse()**: Reverse the list. 




```
x=[1,2,3,4,5]
print(x)
y=[5,6,7,8]
print(y)
x.append(3)
print(x)
x.extend(y)
print(x)
x.pop(0)
print(x)
x.insert(1,0)
print(x)
x.remove(5)
print(x)
x.sort()
print(x)
x.reverse()
print(x)
```

At this point, something that you might have notice is that **methods are actually modifying the list directly**. Once again this is due to the mutable nature of lists, i.e. lists can be changed even after creation. As a result, **lists methods often do not return a value/a new list**.

```
x=[1,2,3,4,5]
x.append(3)
print(x)
another_x=x.append(5)
print(another_x)
```

### 3.1.5 Iterating through a list

A list is said to be iterable, i.e. you can access each element in a list, one by one.

Thus, as you can imagine a list can be used as a counter for loops.

```
x=[1,2,3,4,5]

for number in x:
  print(number)
```

At this point, you will notice that strings and lists are actually very similar. Strings are actually just a ordered sequence of characters, that is indexable, iterable, but immutable. It would be useful to convert a string into a list using the following function, **list()**. Note that **list()** only takes in iterable arguments.

```
x='apple'
print(x)
y=list(x)
print(y)
```

```
scbe_modules={
    'cbe_modules':{
    'Year 1':['MEB', 'Organic Chem', 'Thermodynamics'],
    'Year 2':['Chemical Thermodynamics', 'Fluid Systems'],
    'Year 3':['Process Control', 'Advanced Unit Operations'],
    'Year 4':['Final Year Design Project']    
    },
    'bie_modules':{
    'Year 1':['Anatomy and Physiology', 'Mathematics 1'],
    'Year 2':['Cardiovascular Engineering','Biological Thermodynamics'],
    'Year 3':['Biomedical Imaging','Control in Biosystems'],
    'Year 4':['Final Year Project']
    }    
}

scbe_modules['UEs']=['German Language Level 1', 'Japanese Language Level 1']    # Creates a new key:value pair
print(scbe_modules)

scbe_modules['bie_modules']['Year 2'].append('Computational Methods')    #.append() works since the value of 'Year 2' is a list
print(scbe_modules)
```

### 3.2.3 Dictionary Functions/Operations

* **len(dictionary)**: Returns the number of key:value pairs in the dictionary (outermost)
* element **in** dictionary: Check if the element is a key in the dictionary (outermost)
* **zip(list_1, list_2)**: Creates pairs from two parallel lists. Useful for building dictionaries.
* **dict(lists of pairs)**: Takes  a list of pairs to make a dictionary. e.g.
dict(zip(['fruits','drinks'],['apple','water']))



```
scbe_modules={
    'cbe_modules':{
    'Year 1':['MEB', 'Organic Chem', 'Thermodynamics'],
    'Year 2':['Chemical Thermodynamics', 'Fluid Systems'],
    'Year 3':['Process Control', 'Advanced Unit Operations'],
    'Year 4':['Final Year Design Project']    
    },
    'bie_modules':{
    'Year 1':['Anatomy and Physiology', 'Mathematics 1'],
    'Year 2':['Cardiovascular Engineering','Biological Thermodynamics'],
    'Year 3':['Biomedical Imaging','Control in Biosystems'],
    'Year 4':['Final Year Project']
    }    
}
print(len(scbe_modules))
print('MEB' in scbe_modules)   #Will print False, since in search from the outermost dictionary (i.e. cbe_modules and bie_modules keys)

print(zip(['fruits','drinks'],['apple','water']))   #This will print the address of the zip object

print(dict(zip(['fruits','drinks'],['apple','water'])))
```

### 3.2.4 Dictionary Methods

* dictionary.items(): acess all key:value pairs (outermost). Later on when using dictionary.items() for iteration, the first variable it returns is the key, and the second variable would be the value of the key. 
* dictionary.keys(): access all keys (outermost)
* dictionary.values(): acess all values (outermost)
* dictionary.clear(): Empties the dictionary
* dictionary.update(another_dictionary): For each key:value pair in another_dictionary, update dictionary with that new key:value pair.

This methods will be useful for iteration through a dictionary (next section).



```
scbe_modules={
    'cbe_modules':{
    'Year 1':['MEB', 'Organic Chem', 'Thermodynamics'],
    'Year 2':['Chemical Thermodynamics', 'Fluid Systems'],
    'Year 3':['Process Control', 'Advanced Unit Operations'],
    'Year 4':['Final Year Design Project']    
    },
    'bie_modules':{
    'Year 1':['Anatomy and Physiology', 'Mathematics 1'],
    'Year 2':['Cardiovascular Engineering','Biological Thermodynamics'],
    'Year 3':['Biomedical Imaging','Control in Biosystems'],
    'Year 4':['Final Year Project']
    }    
}

print(scbe_modules.items())
print(scbe_modules.keys())
print(scbe_modules.values())

print(scbe_modules.update({'UEs':['German Language Level 1', 'Japanese Language Level 1']}))   # Since scbe_modules is a mutable, the method changes it directly, does not return a new dictionary

scbe_modules.update({'UEs':['German Language Level 1', 'Japanese Language Level 1']})
print(scbe_modules)
```

### 3.2.5 Iterating through a dictionary

Notice earlier that .items(), .keys() and .values() return a list. Therefore this would be a conveninent way for iterating through the key:value pairs in the diciontary in a **ordered** manner.



```
scbe_modules={
    'cbe_modules':{
    'Year 1':['MEB', 'Organic Chem', 'Thermodynamics'],
    'Year 2':['Chemical Thermodynamics', 'Fluid Systems'],
    'Year 3':['Process Control', 'Advanced Unit Operations'],
    'Year 4':['Final Year Design Project']    
    },
    'bie_modules':{
    'Year 1':['Anatomy and Physiology', 'Mathematics 1'],
    'Year 2':['Cardiovascular Engineering','Biological Thermodynamics'],
    'Year 3':['Biomedical Imaging','Control in Biosystems'],
    'Year 4':['Final Year Project']
    }    
}

for course,years in scbe_modules.items():   #Access the first dictionary. 
  #In this case, course is the key, and years would be the value of the key.
  print(course)
  for year,modules in years.items():        #Access the second dictionary. In this case, years is also a dictionary
    print(year)
    for module in modules:                  #Print out the values of the second dicionary
      print(module)
```

## **Sets**: Acessing, Functions, Methods

In sets, *just like in Math*, it is simply a collection of items which do not have any index or order. This is represented in Python by curly brackets. 

*(Imagine: It's just like a shopping basket where everything is thrown inside. You don't need extras of everything, so duplicates don't exist)*

Take note! You cannot change the items in your set, but you can add new items into your set. 


```
shoppingBasket={'apples','bananas','grapes','grapes'} #if you have duplicates, see what happens when you print

print(shoppingBasket)
```

This means that you can no longer access items in the set with an index. However, you can still check if the item exists in a set. 

```
shoppingBasket={'apples','bananas','grapes'}
print('apples' in shoppingBasket)
```

You can also loop through the items in the set. Just like going through the items in your shopping basket. 

```
shoppingBasket={'apples','bananas','grapes'}
for grocery in shoppingBasket:
  print(grocery)
```

**Adding items to a set**

```
shoppingBasket={'apples','bananas','cherries'}
print(shoppingBasket)
shoppingBasket.add('blueberries')
print(shoppingBasket)
shoppingBasket.update(['lychees','milk'])
print(shoppingBasket)
```

```
shoppingBasket.add('milk')
print(shoppingBasket)
```

**Getting the length of set**

Use the *len()* method, similar to lists and tuples. 

```
shoppingBasket={'apples','bananas','cherries'}
print(len(shoppingBasket))
```

**Removing items in set**

You can use either the *remove()* or *discard()* method to remove items in the set. 

```
shoppingBasket={'apples','bananas','cherries'}
shoppingBasket.remove('apples')
print(shoppingBasket)
shoppingBasket.discard('bananas')
print(shoppingBasket)
```

You can also use the *pop()* method which will remove the last item in the set, but as the set is unordered, you will not know which item is removed. 

```
shoppingBasket={'apples','bananas','cherries'}
shoppingBasket.pop()
print(shoppingBasket)
```

**Clearing the set**

You can also use the *clear()* method to clear the items from the set

```
shoppingBasket={'apples','bananas','cherries'}
shoppingBasket.clear()
print(shoppingBasket)
```

**Deleting the set**

You can also use *del* to delete a set completely. 

```
shoppingBasket={'apples','bananas','cherries'}
del shoppingBasket
print(shoppingBasket)
```

## List and Dictionary Comprehension

List comprehension is simply a concise way of creating lists in python. The general syntax of list comprehension is as follows.

new_list=[expression for item in iterable if condition]

This is equivalent to

new_list=[ ]

for item in iterable:
> if condition:
>> new_list.append(expression)

Similarly, dictionary comprehension is a concise way of creating dictionaries.

new_dict={ key:value for key,value in iterable if condition }

This is equivalent to

new_dict={ }

for key,value in iterable:
> if condition:
>> new_dict[key]=value




```
new_list=[item for item in [1,2,3,4,5] if item!=1]

print(new_list)
```
